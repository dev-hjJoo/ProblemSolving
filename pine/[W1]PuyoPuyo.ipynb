## PuyoPuyo

```python
from collections import deque

stage = []
connectedStone = []
bombQueue = deque()
clear = [False,False,False,False,False,False]

dx = [0,0,-1,1]
dy = [-1,1,0,0]

def findPuyo(stage, x, y,target):
    global connectedStone
    if(connectedStone.count([x,y])==0):
        connectedStone.append([x,y])
    for index in range(4):
        nx = x+dx[index]
        ny = y+dy[index]
        
        if nx<0 or nx>=6 or ny<0 or ny>=len(stage) :
            continue
        
        nearby = stage[ny][nx]
        if(nearby == target and connectedStone.count([nx,ny]) == 0):
            connectedStone.append([nx,ny])
            findPuyo(stage,nx,ny,target)
    
    
    
def bomb(stage):
    global bombQueue
    
    while bombQueue:
        group = bombQueue.pop()
        for area in group:
            x, y = area
            stage[y][x]='.'

def sortStage(stage):
    global bombQueue
    
    for checkX in range(6):
        if(clear[checkX]==False):
            changeArray=[]
            for y in range(len(stage)):
                checkStone = stage[y][checkX]
                if(checkStone!='.'):
                    changeArray.append(checkStone)
                    stage[y][checkX]='.'
            if(len(changeArray)==0):
                clear[checkX]=True
            else:
                for i in range(len(changeArray)):
                    stage[i][checkX]= changeArray[i]
    
for _ in range(12):
    row = list(input())
    if(row != ['.', '.', '.', '.', '.', '.']):
        stage.append(row)

stage.reverse()
count = 0
isPossibleBomb=True

def puyopuyo():
    global connectedStone
    global count
    global isPossibleBomb
    
    target =  ''
    for yIndex in range(len(stage)):
        for index in range(6):
            stone = stage[yIndex][index]
            if(stone !='.'):
                target = stone
                findPuyo(stage,index,yIndex,target)
        
                if(len(connectedStone)>=4):
                    connectedStone.sort()
                    if(bombQueue.count(connectedStone)==0):
                        bombQueue.append(connectedStone)
                    connectedStone=[]
                else:
                    connectedStone=[]
    
    if(len(bombQueue)>0):
        bomb(stage)
        sortStage(stage)
        count += 1
    else:
        isPossibleBomb=False

while isPossibleBomb:
    puyopuyo()
# print(*bombQueue)
print(count)
```
